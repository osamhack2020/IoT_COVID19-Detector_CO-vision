```python
import cv2
import numpy as np
import pytesseract
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("junha.png")

img = img[500:550, 50:200]

img = cv2.resize(img, None, fx=5, fy=5)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
adaptive_threshold = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 85, 7)
cv2.imshow('img',adaptive_threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(12, 10))
plt.imshow(adaptive_threshold, cmap='gray')
```

```python
edges = cv2.Canny(adaptive_threshold,100,200)

cv2.imshow('img',edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_result = edges
chars = pytesseract.image_to_string(img_result, lang='kor', config='--psm 7 --oem 0')
print(chars)
```
